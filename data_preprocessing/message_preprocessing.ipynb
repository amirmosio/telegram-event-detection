```python
dataframe = pd.read_csv(r'C:\Users\marti\OneDrive\Desktop\politecnico\magistrale\telegram-event-detection\trial.csv')
print(dataframe['reactions'])
```

```python
import re
def reactions_splitting(message):
    pattern = r"emoticon=['\"](.*?)['\"].*?count=(\d+)"
    matches = re.findall(pattern, message)
    if matches:
        return matches
    else:
        return None
```

```python
dataframe['reactions'].fillna('', inplace=True)
dataframe['emoji_count'] = dataframe['reactions'].apply(lambda x: reactions_splitting(x))
```

LANGUAGE TRANSLATOR

```python
print(ts.translators_pool)
```

```python
url_pattern = r'https?://\S+'
df_original = dataframe.copy()
df_original = df_original.dropna(subset=['text']) #drop messages composed by csv files
rows_with_links = df_original[df_original['text'].str.contains(url_pattern)] #drop message composed by links
df_original = df_original.drop(rows_with_links.index)
df_original = df_original.reset_index(drop=True) #reset index
df_translated = df_original.copy()
```

```python
translation_text_google = []
translation_text_bing = []

for i in range(len(df_original)):
        message_google = ts.translate_text(query_text = df_original['text'][i], translator='google', from_language='auto', to_language='it')
        message_bing = ts.translate_text(query_text = df_original['text'][i], translator='bing', from_language='auto', to_language='it')
        translation_text_bing.append(message_bing)
        translation_text_google.append(message_google)

df_translated['translation_text_google'] = translation_text_google
df_translated['translation_text_bing'] = translation_text_bing
```

----------------------------------------------------------------------------------------------------------------
REMOVE STOPWORDS

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
print(stopwords.words('english'))
```

```python
stop_words = set(stopwords.words('english'))
texts = df_original['text']
text_tokens = [word_tokenize(text) for text in texts ]
print(text_tokens)
```

```python
filtered_sentences = []
for text in text_tokens:
    filtered_sentence = [w for w in text if not w.lower() in stop_words]
    filtered_sentences.append(filtered_sentence)
print(filtered_sentences)
```

```python
df_translated['tokized'] = filtered_sentences
df_translated.head()
```

----------------------------------------------------------------------------------------------------------------
STEMMING

```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
```

```python
porter_stemmer = PorterStemmer()
stemmer = SnowballStemmer(language='english')

stemmed_sentences, stemmed_sentences2 = [], []
for sentence in filtered_sentences:
    stemmed_words = [porter_stemmer.stem(word) for word in sentence]
    stemmed_words2 = [stemmer.stem(word) for word in sentence]
    stemmed_sentences.append(stemmed_words)
    stemmed_sentences2.append(stemmed_words2)
print(stemmed_sentences)
print(stemmed_sentences2)
```

```python
df_translated['stemmed_sentence_Porter'] = stemmed_sentences
df_translated['stemmed_sentence_snowball'] = stemmed_sentences2

df_translated.head()
```
