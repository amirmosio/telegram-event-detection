```python
import pandas as pd
```

```python
dataframe = pd.read_csv('trial.csv')
```

```python
dataframe['reactions']
```

```python
import re
def reactions_splitting(message):
    pattern = r"emoticon=['\"](.*?)['\"].*?count=(\d+)"
    matches = re.findall(pattern, message)
    if matches:
        return matches
    else:
        return None
```

```python
dataframe['reactions'].fillna('', inplace=True)
dataframe['emoji_count'] = dataframe['reactions'].apply(lambda x: reactions_splitting(x))
```

```python
dataframe['emoji_count']
```

```python
dataframe.to_csv('reaction_trial.csv', index=False)
```

```python
dataframe
```

```python
texts = dataframe['text']
print(texts)
```

```python
langs = ["en" for i in range(len(texts))]
```

```python
%%time


from laserembeddings import Laser

laser = Laser()

# if all sentences are in the same language:

embeddings = laser.embed_sentences(
    texts,
    lang=langs)
```

```python
embeddings.shape
```

```python
print(embeddings[0], texts[0])
```

```python
reference = 'meet'
embedding = laser.embed_sentences(
    reference,
    lang='en')
print(embedding)
```

```python
embeddings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
similarities = cosine_similarity(embedding, embeddings)
```

```python
most_similar = np.argmax(similarities)
most_similar
```
